```python
from itertools import groupby

import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
titanic = pd.read_csv('../data/tested.csv')
titanic.head()
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

ScatterPlot (Numerical - Numerical)


```python
sns.scatterplot(x = tips['total_bill'], y = tips['tip'], hue= tips['sex'], style = tips['smoker'])
```

Bar Plot (Numerical - Categorical)

```python
sns.barplot(x = titanic['Pclass'], y = titanic['Age'], hue= titanic['Sex'])
```

Box Plot (Numerical - Categorical)

```python
sns.boxplot(x = titanic['Sex'], y = titanic['Age'], hue= titanic['Survived'])
```

Distplot (Numerical - Categorical)

```python
sns.displot(x = titanic[titanic['Survived']==0]['Age'], kde=True)
sns.displot(x = titanic[titanic['Survived']==1]['Age'], kde=True)
```

HeatMap (Categorical - Categorical)

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
titanic.groupby('Pclass')['Survived'].mean()*100
```

Pairplot

```python
sns.pairplot(iris, hue='species')
```
